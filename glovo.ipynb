```python
import pandas as pd

segments_df = pd.read_csv('data_business_segments.csv')
segments_df
countries_df = pd.read_csv('data_countries.csv')
countries_df
finance_df = pd.read_csv('data_finance.csv')
finance_df
integrations_df = pd.read_csv('data_integrations.csv')
integrations_df
operations_df = pd.read_csv('data_operations.csv')
operations_df
orders_df = pd.read_csv('data_orders.csv')
orders_df
vertical_df = pd.read_csv('data_vertical.csv')
vertical_df
```

```python
print(vertical_df.shape, orders_df.shape,operations_df.shape,integrations_df.shape,finance_df.shape,countries_df.shape,segments_df.shape)
```

```python
dimentions_df = pd.merge(integrations_df, countries_df, on=['partner_id']).merge(segments_df, on=['partner_id']).merge(vertical_df, on=['partner_id'])
```

```python
dimentions_df.dropna(axis='index', how='any')
```

```python
daily_metrics_df = orders_df.merge(finance_df, on=['partner_id','date']).merge(operations_df, on=['partner_id','date'])
```

```python
daily_metrics_df.dropna(axis='index', how='any')
```

```python
grouped_metrics_df = daily_metrics_df.groupby('partner_id').agg({'orders_daily':'sum',
                                            'connected_hours': 'sum',
                                            'avg_order_cost': 'mean',
                                            'avg_order_revenue': 'mean',
                                            'avg_delivery_time_min': 'mean',
                                            'avg_preparation_time_min': 'mean'}).reset_index()
```

```python
final_df = dimentions_df.merge(grouped_metrics_df, on=['partner_id'])
```

```python
final_df.to_csv('glovo_data.csv', index=False)
```
